```python
import torch
from transformers import BertModel, BertTokenizer
# 这里我们调用bert-base模型，同时模型的词典经过小写处理
model_name = 'bert-base-chinese'
# 读取模型对应的tokenizer
tokenizer = BertTokenizer.from_pretrained(model_name)
# 载入模型
model = BertModel.from_pretrained(model_name)
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv('./training/label_meme.csv')

print(df.info())
print(df.head())
```

```python
com = df["context"].values
file =  df["id"].values

print(com.shape)
comment = []
```

```python

for i in range(5169):
    # 通过tokenizer把文本变成 token_id
    print(file[i])
    if type(com[i])==float:
      last_hidden_states = np.zeros([1,768], dtype = float) 
      last_hidden_states = torch.from_numpy(last_hidden_states)
      # print(torch.mean(last_hidden_states,0))
    else:
      com[i] = com[i].replace('\n',' ')
      
      input_ids = tokenizer.encode(com[i], add_special_tokens=True)

    
      # input_ids: [101, 2182, 2003, 2070, 3793, 2000, 4372, 16044, 102]
      input_ids = torch.tensor([input_ids])
      with torch.no_grad():
        last_hidden_states = model(input_ids)[0]  # Models outputs are now tuples
      # print(input_ids) 
      last_hidden_states = last_hidden_states.numpy()
      # print(last_hidden_states)
      last_hidden_states = np.reshape(last_hidden_states, (last_hidden_states.shape[1], last_hidden_states.shape[2]))
      # print(last_hidden_states.shape)
      last_hidden_states = torch.from_numpy(last_hidden_states)
      
      # print(last_hidden_states)
      # print(last_hidden_states.size())
      # print(torch.mean(last_hidden_states,0))
   
    np.save('./context_bert/'+file[i], torch.mean(last_hidden_states,0))
   
```
